##### Q2. Write a program that prompts the user to enter their full name then prints the number of characters in their name (do not count spaces).

```
What is your name? Amulya Katti
Amulya Katti your name has 11 characters.
```

````
# Type solution here
# Ask for user input
#Assue characters include letters, numbers, and special characters
user_input = input("What is your name? ")
user_input = user_input.strip()
# List of user name
charList = list(user_input)
counter = 0
# Loop over each character
for char in charList:
  if char != " ":
    counter += 1
# Display output
print(user_input + " your name has " + str(counter) + " characters.")
````

Enter the month number: 9

##### Q4. Write a program that prompts the user to enter a loan amount, annual interest rate, and number of years  for a car loan. Then calculate and print the monthly payment amount.

Note:
```
PMT= [PV*i*((1+i)^n)]/((1+i)^n)-1
```
PMT is the monthly payment

PV is the loan amount

i is the interest rate per month in decimal form (interest rate percentage divided by 12)

n is the number of months (term of the loan in months)

```
Loan Amount: 30000.00
Annual Interest Rate: 4.00
Years: 5
Your monthly payment is: $552.50
```

````
# Type solution here
# Ask for user input
pv = float(input("Loan Amount: "))
intRate = float(input("Annual Interest Rate: "))
yrs = float(input("Years: "))
# Percentage interest rate and number of months
i = (intRate / 12) * 0.01
n = yrs * 12
# Calculation of monthly payment
pmt = (pv * i * ((1 + i)** n))/(((1+i)** n) - 1)
# Display output
print("Your monthly payment is: $" + "{:0.2f}".format(pmt))
````

##### Q5. Write a program to that prompts the user to enter a string. Then check whether or not the string is a palindrome.

```
Enter a string: Madam
Madam, is a palindrome!
```

````
# Type solution here
# Ask for user input
userInput = input("Enter a string: ")
uInput = userInput.strip().lower()
# Reverse user input
revInput = uInput[::-1]
# Is palindrome
if revInput == uInput:
  print(userInput + ", is a palinedrome!")
# Not palindrome
if revInput != uInput:
  print(userInput + ", is not a palinedrome.")
````

##### Q6. Write a program to compute and print all possible combinations of change for \$1. Denominations to be considered – quarter, dime, nickel, penny.

```
Change for $1:
0 quarters, 0 dimes, 0 nickels, 100 pennies
…
4 quarters, 0 dimes, 0 nickels, 0 pennies
```

````
# Type solution here
print("Change for $1:")
# Loop over each type of coins
for qNum in range(0,6):
  for dNum in range(0, 11):
    for nNum in range(0, 21):
      for pNum in range(0, 101):
        totalCoins = qNum * 0.25 + dNum * 0.1 + nNum * 0.05 + pNum * 0.01
        # Display the combination if add up to $1
        if totalCoins == 1:
           print(str(qNum) + " quarters, " + str(dNum) + " dimes, ", str(nNum) 
                  + " nickels, " + str(pNum) + " pennies")
````

##### Q7. Ask the user to enter two positive integers between 1 and 100. Read those integers. Then output a multiplication table of the first number times the second number. 

Note: You have to ensure the entered numbers are between 1 and 100 only.

```
Please enter an integer: 5
Please enter another integer: 20
5 x 1 = 5
5 x 2 = 10
5 x 3 = 15
…
5 x 20 = 100
```

````
# Type solution here
# First input
uInput_1 = input("Please enter an integer: ")
# Invalid input
while not (uInput_1.isdigit() and int(uInput_1) in range(1, 101)):
  uInput_1 = input("Please enter an integer: ")
# Second input
uInput_2 = input("Please enter another integer: ")
# Invalid input
while not (uInput_2.isdigit() and int(uInput_2) in range(1, 101)):
  uInput_2 = input("Please enter another integer: ")
# Loop over the second user input 
for i in range(1, int(uInput_2) + 1):
  prod = int(uInput_1) * i
  print(uInput_1 + " x " + str(i) + " = " + str(prod))
````

##### Q8. Write a program to ask the user to enter a password. Then check to see if it is a valid password based on these requirements -

Must be at least 8 characters long

Must contain both uppercase and lowercase letters

Must contain at least one number between 0-9

Must contain a special character -!,@,#,$

If the password is not valid, ask the user to re-enter. This should continue until the user enters a valid password. After a valid password is entered, print Access Granted!

```
Please enter a password. Follow these requirements - 
a. Must be at least 8 characters long
b. Must contain both uppercase and lowercase letters
c. Must contain at least one number between 0-9
d. Must contain a special character -!,@,#,$
Password: HelloWorld1
Invalid password. Try again!
Password: Hello@World1
Access Granted!
```


````
# Type solution here
print("Please enter a password. Follow these requirements")
print("a. Must be at least 8 characters long")
print("b. Must contain both uppercase and lowercase letters")
print("c. Must contain at least one number between 0-9")
print("d. Must contain a special character -!,@,#,$")
while True:
    # Ask for user input
    password = input("Password: ")
    # Dictionary that checks if user input is valid
    pwDict = {"upper": 0, "lower": 0, "num": 0, "spchar": 0}
    # Loop over user input and store them into pwDict
    if len(password) >= 8:
        for char in password:
            if char.isupper():
                pwDict["upper"] += 1
            if char.islower():
                pwDict["lower"] += 1
            if char in "123456789":
                pwDict["num"] += 1
            if char in "!@#$":
                pwDict["spchar"] += 1
    # If input is valid
    validInput = True     
    # Loop over dictionary
    for value in pwDict:
        # Invalid input
        if pwDict[value] == 0:
            print("Invalid password. Try again!")
            validInput = False
            break
    # Valid input
    if validInput == True:
        break            
print("Access Granted!")
````

##### Q9. Given the score sheet of participants in a competition, you are required to find the runner-up score. You are given scores in a list. Find the score of the runner-up.

```
list1 - [2, 3, 6, 6, 5]
runner_up - 5

list2 - [1, 7, 5, 3, 10, 4, 5, 5, 6, 11]
runner_up - 10
```

````
# Type solution here
# Given list here
lst = [2, 3, 6, 6, 5]
# New list that eliminates repeated numbers
newList = []
# Append numbers to new list
for num in lst:
  if num not in newList:
    newList.append(num)
# Sort numbers in reverse order
newList.sort(reverse = True)
# Runner up is in the second place
print("runner_up - " + str(newList[1]))
````
